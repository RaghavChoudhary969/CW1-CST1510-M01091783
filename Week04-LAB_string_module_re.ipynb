4. Create a function (or set of functions ) taking one parameter for a password.

The function (s) will print True if

  - the password contains at least one upper case letter,   
  - the password contains at least one lower case letter,
  - the password contains at least one digit and one special character like for example ‚Äò@‚Äô.
- Extra (check if the password contains space. If so, print a suitable error message)

````python
def validate_password(password):
    if " " in password:
        print("‚ùå Password should not contain spaces.")
        return False

    has_upper = any(ch.isupper() for ch in password)
    has_lower = any(ch.islower() for ch in password)
    has_digit = any(ch.isdigit() for ch in password)
    has_special = any(not ch.isalnum() for ch in password)

    if has_upper and has_lower and has_digit and has_special:
        return True
    else:
        return False


password = input("Enter your password: ")
print("‚úÖ Valid Password") if validate_password(password) else print("‚ùå Invalid Password")
````

10. Create a function (s) counting all lower case, upper case, digits, and special symbols from a given string and print them out.
   
```python

   example_password = "P@#yn26at^&i5ve"
```

````python
def count_characters(s):
    lower = upper = digits = special = 0

    for char in s:
        if char.islower():
            lower += 1
        elif char.isupper():
            upper += 1
        elif char.isdigit():
            digits += 1
        else:
            special += 1

    print("Lowercase letters:", lower)
    print("Uppercase letters:", upper)
    print("Digits:", digits)
    print("Special symbols:", special)

example_password = "P@#yn26at^&i5ve"
count_characters(example_password)
````

````python
def count_characters(s: str):
    lower = upper = digits = special = 0

    for ch in s:
        if ch.islower():
            lower += 1
        elif ch.isupper():
            upper += 1
        elif ch.isdigit():
            digits += 1
        else:
            special += 1

    print(f"\nPassword: {s}")
    print("Lowercase letters:", lower)
    print("Uppercase letters:", upper)
    print("Digits:", digits)
    print("Special symbols:", special)



password_list = ["P@#yn26at^&i5ve", "Hello123!", "passWORD2025$", "abcDEF", "1234@#"]


for password in password_list:
    count_characters(password)
````

12. Password Validation with Multiple Criteria: Ask the user for a password and validate whether it:
- Has more than 8 characters.
- Contains at least one uppercase letter.
- Contains at least one lowercase letter.
- Contains at least one number.
- Contains at least one special character (@, #, $, etc.).
- Extra: If the password contains a space, print an error message

````python
def validate_password(password: str):
    # Criteria checks
    length_ok = len(password) > 8
    upper_ok = any(ch.isupper() for ch in password)
    lower_ok = any(ch.islower() for ch in password)
    digit_ok = any(ch.isdigit() for ch in password)
    special_ok = any(ch in "@#$%^&*!?" for ch in password)
    space_error = " " in password


    print(f"\nPassword entered: {password}")
    if space_error:
        print("‚ùå Error: Password should not contain spaces.")
        return

    if all([length_ok, upper_ok, lower_ok, digit_ok, special_ok]):
        print("‚úÖ Password is strong and valid!")
    else:
        print("‚ùå Password is invalid. Please check the following:")
        if not length_ok:
            print("- Must be longer than 8 characters")
        if not upper_ok:
            print("- Must contain at least one uppercase letter")
        if not lower_ok:
            print("- Must contain at least one lowercase letter")
        if not digit_ok:
            print("- Must contain at least one number")
        if not special_ok:
            print("- Must contain at least one special character (@, #, $, %, ^, &, *, !, ?)")


# Ask user for input
user_password = input("Enter your password: ")
validate_password(user_password)
````
